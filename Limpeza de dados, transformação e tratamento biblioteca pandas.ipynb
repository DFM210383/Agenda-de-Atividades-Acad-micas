# 6. Operações e Transformações de Dados com Pandas

```python
# criando coluna

df_dados['Salário Anual'] = df_dados['Salário'] * 12
df_dados
```

```python
# Funções Lambda (Anônimas)

df_dados['Bônus'] = df_dados['Salário'].apply(lambda x: x * 0.10 if x > 7000 else x * 0.05)
df_dados
```

```python
#Definir os critérios

condicoes = [

             [df_dados'Idade'] < 18,
             (df_dados['Idade'] >= 18) & (df_dados['Idade']<= 30),
             (df_dados['Idade'] > 30) & (df_dados['Idade']<= 60),
             df_dados['Idade'] > 60

]
```

```python
# Rótulos

Faixa = ["Menor de Idade", "Jovem", "Adulto", "Idoso"]
```

```python
df_dados["Faixa Etaria"] = np.select(condicoes,Faixa,default='Idade não informada')
df_dados
```

# 7. Agrupamento de Dados (Group By) com Pandas

```python
# Média salaria por cidade

media_salario = df_dados.groupby('Cidade')['Salário'].mean()
print(np.round(media_salario,2))
```

```python
# multiplas agregações

agregacao = df_dados.groupby('Cidade').agg(Media_salarial = ('Salário','mean'),
                                           Idade_maxima = ('Idade', 'max'),
                                           Contagem = ('Nome','count')

                                           )
```

```python
agregacao
```

# 8. Criação de Tabelas Dinâmicas (Pivot Tables) com Pandas

```python
dados_vendas = {
    'Data': pd.to_datetime(['2026-09-01', '2026-09-01', '2026-09-02', '2026-09-02', '2026-09-01']),
    'Região': ['Norte', 'Sul', 'Norte', 'Sul', 'Norte'],
    'Vendedor': ['Carlos', 'Ana', 'Carlos', 'Ana', 'Pedro'],
    'Vendas': [250, 300, 150, 400, 200]
}
```

```python
vendas = pd.DataFrame(dados_vendas)
vendas
```

```python
tabela_dinamica = vendas.pivot_table(values = 'Vendas',
                                     index = 'Região',
                                     columns= 'Vendedor',
                                     aggfunc= 'sum',
                                     fill_value = 0
                                     )
```

```python
tabela_dinamica
```
