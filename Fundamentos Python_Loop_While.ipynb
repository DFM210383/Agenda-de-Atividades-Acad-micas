# **Loops no Python** *(Laços de Repetição)*

Depois de conhecermos os comandos estruturados de decisão e escolha, veremos, agora, as estruturas de repetição. Estes comandos são conhecidos como Looping e Laços. Este apelido se dá pelo fato dele iniciar e terminar no mesmo ponto e a sua “volta” envolver o bloco de repetição.

# **LAÇO ENQUANTO-FAÇA (LOOP WHILE)**

```python
# VOCÊ CONCORDA COM A SEGUINTE AFIRMAÇÃO?

enquanto há algo para fazer
    faça isso
```

O laço while é usado para repetir uma sequência de instruções, enquanto uma determinada condição for verdadeira. A sintaxe do while loop em Python é a seguinte:

```python
# COMO É NO PYTHON

while <condição>:
  <Bloco de repetição>
```

```python
while conditional_expression:
    instruction_one
    instruction_two
    instruction_three
    :
    :
    instruction_n
```

**PONTOS IMPORTANTES**


*   se você quiser executar mais de uma instrução dentro de um loop while, você deve (como no if) **recuar todas as instruções** da mesma maneira;
*   uma instrução ou um conjunto de instruções executadas dentro do loop while é chamado de **corpo do loop**;
*   se a condição for False (igual a zero) quando for testada pela primeira vez, o **corpo não será executado uma única ve**z (observe a analogia de não ter que fazer nada se não houver nada a fazer);
*   o corpo deve ser capaz de **alterar o valor da condição**, porque se a condição for True no início, o corpo poderá ser executado continuamente até o infinito (observe que fazer algo geralmente diminui o número de coisas a fazer).









```python
#LOOP INFINITO

while 5 == 5:
    print("Estou preso dentro de um loop.")
```

```python
# Usando uma variável "counter" para sair do loop

counter = 10
while counter != 0:
    print("Dentro do laço.", counter)
    counter -= 1
print("Fora do circuito.", counter)
```

```python
# Usando o loop while para imprimir os valores de 0 a 10
# A condição tem que deixar de ser verdadeira dentro do loop, senão pode travar o navegador ou mesmo o computador

valor = 0
while valor < 11:
    print(valor)
    valor = valor + 1
```

```python
# Inicializa a variável 'valor' com 0. Este será o ponto de partida para a contagem.
valor = 0

# Inicia um loop 'while' que continuará executando enquanto o valor de 'valor' for menor que 11.
while valor < 11:
    # Imprime o valor atual da variável 'valor'.
    print(valor)
    # Incrementa o valor de 'valor' em 1 a cada iteração do loop.
    # Isso é crucial para que a condição do 'while' eventualmente se torne falsa e o loop termine.
    valor = valor + 1
```

```python
# Calculadora de Índice de Massa Corporal (IMC) com verificação de peso normal

# Imprime um cabeçalho para a calculadora de IMC
print("\n--- Calculadora de IMC ---")

# Loop para garantir que o peso seja inserido corretamente. Ele se repete até uma entrada válida ser fornecida.
while True:
    try:
        # Solicita ao usuário que digite o peso em quilogramas e tenta converter para float.
        peso = float(input("Digite seu peso em kg (ex: 70.5): "))
        # Verifica se o peso é um valor positivo.
        if peso <= 0:
            print("O peso deve ser um valor positivo. Tente novamente.")
        else:
            # Sai do loop se o peso for válido.
            break
    except ValueError:
        # Captura erros se a entrada não for um número e solicita nova entrada.
        print("Entrada inválida para o peso. Por favor, digite um número.")

# Loop para garantir que a altura seja inserida corretamente. Ele se repete até uma entrada válida ser fornecida.
while True:
    try:
        # Solicita ao usuário que digite a altura em metros e tenta converter para float.
        altura = float(input("Digite sua altura em metros (ex: 1.75): "))
        # Verifica se a altura é um valor positivo.
        if altura <= 0:
            print("A altura deve ser um valor positivo. Tente novamente.")
        else:
            # Sai do loop se a altura for válida.
            break
    except ValueError:
        # Captura erros se a entrada não for um número e solicita nova entrada.
        print("Entrada inválida para a altura. Por favor, digite um número.")

# Calcula o Índice de Massa Corporal (IMC) usando a fórmula: peso / (altura * altura).
imc = peso / (altura ** 2)

# Imprime o valor do IMC calculado, formatado com duas casas decimais.
print(f"\nSeu IMC é: {imc:.2f}")

# Inicia a classificação do IMC com base nos valores padrão.
if imc < 18.5:
    # Se o IMC for menor que 18.5, classifica como 'Abaixo do peso'.
    print("Classificação: Abaixo do peso")
    print("Você NÃO está dentro da faixa de peso normal.")
elif 18.5 <= imc < 24.9:
    # Se o IMC estiver entre 18.5 e 24.9, classifica como 'Peso normal'.
    print("Classificação: Peso normal")
    print("Parabéns! Você está dentro da faixa de peso normal.")
elif 25 <= imc < 29.9:
    # Se o IMC estiver entre 25 e 29.9, classifica como 'Sobrepeso'.
    print("Classificação: Sobrepeso")
    print("Você NÃO está dentro da faixa de peso normal.")
elif 30 <= imc < 34.9:
    # Se o IMC estiver entre 30 e 34.9, classifica como 'Obesidade Grau I'.
    print("Classificação: Obesidade Grau I")
    print("Você NÃO está dentro da faixa de peso normal.")
elif 35 <= imc < 39.9:
    # Se o IMC estiver entre 35 e 39.9, classifica como 'Obesidade Grau II'.
    print("Classificação: Obesidade Grau II")
    print("Você NÃO está dentro da faixa de peso normal.")
else:
    # Se o IMC for 40 ou mais, classifica como 'Obesidade Grau III (Mórbida)'.
    print("Classificação: Obesidade Grau III (Mórbida)")
    print("Você NÃO está dentro da faixa de peso normal.")
```

```python
# Exemplo 1: Contagem regressiva simples
print("\n--- Exemplo de Contagem Regressiva ---")
contador = 5
# O loop continua enquanto 'contador' for maior que 0
while contador > 0:
    print(contador)
    contador -= 1 # Decrementa o contador a cada iteração
print("Fim da contagem regressiva!")
```

```python
# Exemplo 2: Pedindo entrada ao usuário até que um número par seja inserido
print("\n--- Exemplo de Validação de Entrada ---")
numero_valido = False
# O loop continua enquanto 'numero_valido' for Falso
while not numero_valido:
    try:
        entrada = int(input("Por favor, digite um número par: "))
        if entrada % 2 == 0:
            print(f"Você digitou um número par: {entrada}. Obrigado!")
            numero_valido = True # Define como Verdadeiro para sair do loop
        else:
            print("Isso não é um número par. Tente novamente.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")
```

```python
# Exemplo 3: Soma acumulada até que a soma exceda 10
print("\n--- Exemplo de Soma Acumulada ---")
soma_total = 0
valor_atual = 0
# O loop continua enquanto 'soma_total' for menor ou igual a 10
while soma_total <= 10:
    valor_atual += 1 # Incrementa um valor a cada iteração
    soma_total += valor_atual # Adiciona o valor atual à soma total
    print(f"Adicionando {valor_atual}, soma atual é {soma_total}")
print(f"A soma final excedeu 10: {soma_total}")
```

```python
print("Digite 0 para finalizar") # Imprime uma mensagem inicial para o usuário, informando como ele pode parar de inserir números.

# Inicializa a variável 'soma' com 0. Esta variável será usada para acumular a soma de todos os números que o usuário digitar.
soma = 0
# Inicializa a variável 'num' com 1. O valor 1 é escolhido para garantir que o loop 'while' seja executado pelo menos uma vez, já que a condição do loop é 'num != 0'.
num = 1

# Início do loop 'while'. Este loop continuará a executar seu bloco de código enquanto o valor de 'num' for diferente de 0.
while num != 0:
  # Bloco de repetição: Estas linhas são executadas a cada iteração do loop.
  # Solicita ao usuário que digite um número, converte a entrada para um número de ponto flutuante e armazena em 'num'.
  num = float(input("Digite um número: "))
  # Adiciona o número recém-digitado ('num') ao valor atual da 'soma'. Isso acumula os números digitados.
  soma = soma + num

# Fluxo depois do final do laço: Quando o usuário digita 0, o loop 'while' termina e a execução continua a partir desta linha.
# Exibe o valor final da 'soma' de todos os números (exceto o 0 que encerrou o loop) digitados pelo usuário.
print("Somatoria = ", soma)
```

**EXEMPLO:** Fazer um programa que encontrar o maior número de um grande conjunto de dados inseridos.

```python
# Armazene o maior número atual aqui.
largest_number = 0

# Insira o primeiro valor.
number = int(input("Digite um número ou digite -1 para parar: "))

# Se o número não for igual a -1, continue.
while number != -1:
    # O número é maior que o maior_número?
    if number > largest_number:
        # Sim, atualize o maior_número.
        largest_number = number
    # Insira o próximo número.
    number = int(input("Digite um número ou digite -1 para parar: "))

# Imprima o maior número.
print("O maior número é:", largest_number)
```

**COMANDOS BREAK, CONTINUE E PASS**

Até agora, tratamos o corpo do loop como uma sequência indivisível e inseparável de instruções que são executadas completamente a cada turno do loop. No entanto, como desenvolvedor, você pode se deparar com as seguintes opções:



*   pode ser desnecessário executar o loop como um todo; você deve se abster de executar o corpo do loop e ir além;
*   pode ser necessário iniciar a próxima curva do loop sem concluir a execução da curva atual.

O Python fornece duas instruções especiais para a implementação dessas duas tarefas. Essas adições, que não melhoram o poder expressivo da linguagem, mas simplificam o trabalho do desenvolvedor,são chamadas de **açúcar sintático**.

*  **break** - sai do loop imediatamente e termina incondicionalmente a operação do loop; o programa começa a executar a instrução mais próxima após o corpo do loop;

*  **pass** - é uma palavra que deve ser usada sempre que o programa requisitar sintaticamente que se preencha uma lacuna, como é o caso da definição de uma função: após a linha do def tem que haver algum conteúdo.

```python
# Se encontramos o número 4 interrompemos o loop

valor = 0
while valor < 10:
    if valor == 4:
        break
    else:
      pass
    print(valor)
    valor = valor + 1
```

*   **continue** - se comporta como se o programa tivesse chegado ao fim do corpo; o próximo turno é iniciado e a expressão de condição é testada imediatamente.

```python
for item in range(0,9):
    if item == 4:
        continue
    print(item)
```

**EXERCÍCIO 4:** Faça um progama em python que lê uma sequência de números inteiros fornecidos pelo usuário e conta quantos números são pares e quantos são ímpares.O programa termina quando zero é digitado

```python
# Inicializa a variável 'n_impar' para contar os números ímpares, começando em zero.
n_impar = 0
# Inicializa a variável 'n_par' para contar os números pares, começando em zero.
n_par = 0

# Solicita ao usuário o primeiro número inteiro. A mensagem informa como parar o programa.
numero = int(input("Digite um número ou digite 0 para parar: "))

# Inicia um loop 'while' que continuará executando enquanto o número digitado for diferente de 0.
# Digitar 0 é a condição para terminar a execução do loop.
while numero != 0:
    # Verifica se o número atual é par usando o operador de módulo (%). Se o resto da divisão por 2 for 0, é par.
    if numero % 2 == 0:
        # Se for par, incrementa o contador de números pares.
        n_par += 1
    else:
        # Se não for par (ou seja, ímpar), incrementa o contador de números ímpares.
        n_impar += 1
    # Após processar o número, solicita ao usuário o próximo número para continuar ou sair do loop.
    numero = int(input("Digite um número ou digite 0 para parar: "))

# Após o loop terminar (quando o usuário digita 0), imprime os resultados finais.
# Mostra a quantidade total de números ímpares digitados.
print("Números ímpares contam:", n_impar)
# Mostra a quantidade total de números pares digitados.
print("Números pares contam:", n_par)
```

**EXERCÍCIO 5:** Em uma eleição, há três candidatos: 1 – Mickey; 2 – Donald e 3 – Pateta.

Fazer um programa que leia os votos dados pelo usuário. Se o usuário digitar 0 (zero) encerra a vontação.
Ao finalizar a votação, informar quantos votos cada candidato obteve.

Ex:

1 – Mickey : 4 votos

2 – Donald: 3 votos

3 – Pateta: 1 voto

```python
# zera as variáveis dos candidatos
hug = 0
zez = 0
lui = 0

# Exibe o menu com os candidatos
print("Digite o voto ou 0 para finalizar")
print("1 – Huguinho")
print("2 – Zezinho")
print("3 – Luizinho")

# inicia um laço infinito
while True:
  # Lê o voto
  voto = int(input("Digite o voto: "))

  # Contabiliza o voto
  if voto == 1:
    hug = hug + 1
  elif voto == 2:
    zez = zez + 1
  elif voto == 3:
    lui = lui + 1
  else:
    if voto != 0:
      print("Voto inválido, digite: 1, 2 ou 3")

  # Este if simula a condição de saída no final do laço
  if voto == 0:
    break # força a saída do laço

# Exibe a quantidade de votos de cada candidato
print(f"1 – Huguinho: {hug} votos")
print(f"2 – Zezinho: {zez} votos")
print(f"3 – Luizinho: {lui} votos")
```
