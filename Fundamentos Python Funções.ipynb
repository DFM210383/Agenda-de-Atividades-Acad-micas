***COMO AS FUNÇÕES SE COMUNICAM O SEU AMBIENTE?***

--> FUNÇÕES PARAMETRIZADAS

O poder total da função se revela quando pode ser equipado com uma interface capaz de aceitar **dados fornecidos pelo invocador**. Esses dados podem modificar o comportamento da função, tornando-a mais flexível e adaptável às condições variáveis

```python
# ESTRUTURA DE UMA FUNÇÃO

def function_name(<parametros>):
   <function_body>
```

*   parâmetros existem apenas dentro de funções nas quais eles foram definidos, e o único lugar onde o parâmetro pode ser definido é um espaço entre um par de parênteses na declaração def;
*   a atribuição de um valor ao parâmetro é feita no momento da chamada da função especificando o argumento correspondente




```python
def mensagem2(number):
 print("Digite um número:", number)

#mensagem2()
mensagem2(1)
```

***Efeitos e resultados: a instrução return***

Todas as funções apresentadas anteriormente têm algum tipo de efeito - elas produzem algum texto e o enviam para o console.

Obviamente, funções podem ter **resultados**.

Para que as funções retornem um valor (mas não apenas para essa finalidade), use a instrução **return**.

```python
def myfunction():
  return 3+3
  #print("Olá, Mundo!")

print(myfunction())
```

**EFEITOS DO RETURN**

*   causa o término imediato da execução da função (nada de novo se comparado à primeira variante)
*   além disso, a função avaliará o valor da expressão e o retornará (daí o nome mais uma vez) como o resultado da função.

***Efeitos e resultados: listas e funções***

uma lista pode ser enviada para uma função como argumento?



```python
def list_sum(lst):
    s = 0
    for elem in lst:
        s += elem
    return s

print(list_sum([5, 4, 3]))

#print(list_sum(5)) erro
```

ESCOPO NAS FUNÇÕES

uma variável **existente fora de uma função tem escopo dentro do corpo da função**

```python
def my_function():
 print("Eu conheço aquela variável?", var)


var = 1
my_function()
print(var)
```

Vamos fazer uma pequena alteração no código:

a variável var criada **dentro da função não é a mesma que quando definida fora dela** - parece que há duas variáveis diferentes com o mesmo nome;

além disso, a variável da função oculta a variável que vem do mundo exterior.

USAMOS A PALAVRA **GLOBAL**

```python
def my_function():
    global var2
    var2 = 2
    print("Eu conheço aquela variável?", var2)


var2 = 1
my_function()
print(var2)
```

**EXEMPLO: Gerador de Senhas**

Descrição: Crie uma função chamada **gerar_senha**

O usuário deve escolher:


*   Quantos carcateres deseja
*   Se deseja letra minúscula
*   Se deseja letra maiúscula
*   Se deseja caracteres especiais


Regras:

A função deve retornar a senha gerada como uma string.

A senha deve ter exatamente o número de caracteres definido por tamanho.

O uso de bibliotecas como random é permitido (e recomendado).

```python
# VERSÃO 1 - PASSANDO OS ARGUMENTOS DIRETAMENTE NA FUNÇÃO

import random

def gerar_senha(tamanho, minusculas=False, maiusculas=False, simbolos=False):
    caracteres = ''

    if minusculas:
        caracteres += 'abcdefghijklmnopqrstuvwxyz'
    if maiusculas:
        caracteres += 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    if simbolos:
        caracteres += '!@#$%&*'

    if not caracteres:
        return "Erro: nenhum tipo de caractere selecionado"

    senha = ''.join(random.choices(caracteres, k=tamanho))
    return senha


# Execução
gerar_senha(10,True,True,True)
```

```python
#VERSÃO 2 - INTERAÇÃO COM O USUÁRIO

import random

def gerar_senha(tamanho, minusculas, maiusculas, simbolos):
    print(" \n === Gerador de Senha ===")

    caracteres = ''

    if minusculas:
        caracteres += 'abcdefghijklmnopqrstuvwxyz'
    if maiusculas:
        caracteres += 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    if simbolos:
        caracteres += '!@#$%&*'

    if not caracteres:
        return "Erro: nenhum tipo de caractere selecionado"

    senha = ''.join(random.choices(caracteres, k=tamanho))
    return senha



# Execução
tamanho = int(input("Quantos caracteres deseja"))
minuscula = input("Deseja letra minuscula[S/N]: ").lower()[0] == 's'
maiscula = input("Deseja letra maiuscula[S/N]: ").lower()[0] == 's'
simbolo = input("Deseja simbolos [S/N]: ").lower()[0] == 's'

gerar_senha(tamanho,minuscula,maiscula,simbolo)
```

**Exercício 4: Ano bissexto**

Sua tarefa é escrever e testar uma função que usa um argumento (um ano) e retorna True se o ano for um ano bissexto ou False caso contrário.

A semente da função já é mostrada no código esqueleto do editor.

Nota: também preparei um código de teste curto, que você pode usar para testar sua função.

O código usa duas listas - uma com os dados de teste e a outra contendo os resultados esperados. O código informará se algum dos resultados é inválido.

```python
def ano_bissexto(ano):

# Solicita o ano ao usuário
#ano = int(input("Digite o ano: "))

  # Verifica se o ano está dentro do período do calendário gregoriano
  if ano < 1582:
      print("Não está dentro do período do calendário gregoriano.")
  else:
      # Verifica se é um ano bissexto ou comum
      if ano % 4 != 0:
          return False
      elif ano % 100 != 0:
          return True
      elif ano % 400 != 0:
          return False
      else:
          return True

teste_data = [1900, 2000, 2016, 1987]
teste_result = [False, True, True, False]

for i in range(len(teste_data)):
 yr = teste_data[i]
 print(yr,"->",end="")
 result = ano_bissexto(yr)

 if result == teste_result[i]:
  print("OK")
 else:
  print("Não deu certo")
```
