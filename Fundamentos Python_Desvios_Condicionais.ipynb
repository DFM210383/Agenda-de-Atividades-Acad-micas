# **CONECTIVOS LÓGICOS**

Os operadores lógicos funcionam assim:

- Operador **and** - Retorna True se ambas as declarações forem verdadeiras.
- Operador **or** - Retorna True se uma das declarações for verdadeira.
- Operador **not** - Inverte o resultado, retorna False se o resultado for True.

```python
# Operador and

numero = 9

if (numero > 2) and (numero % 2 == 0):
    print("Isso está sendo impresso porque as duas condições são verdadeiras!")
```

```python
# Operador and

numero = 9

if (numero > 5) and (numero % 2 == 0):
    print("Isso está sendo impresso porque as duas condições são verdadeiras!")
else:
    print("Isso está sendo impresso porque uma das duas condições é falsa!")
```

```python
# Operador or (OU)

numero = 4

if (numero > 5) or (numero % 2 == 0):
    print("Isso está sendo impresso porque uma duas condições é verdadeira!")
```

```python
# Operador not

numero = 4

if not(numero > 5) and (numero % 2 == 0):
    print("Isso está sendo impresso porque as duas condições são verdadeiras!")
else:
    print("Isso está sendo impresso porque uma das duas condições é falsa!")
```

```python
#Crie um conversor de moeda, pedindo pro usuario digitar o numero referente a moeda e fazendo a conversão em real.
```

```python
# Taxas de câmbio fixas (você pode atualizar essas taxas ou buscar de uma API)
taxas_de_cambio = {
    "dolar": 5.20, # Taxa de câmbio para Dólar Americano
    "euro": 6.10,  # Taxa de câmbio para Euro
    "libra": 7.20, # Taxa de câmbio para Libra Esterlina
    "iene": 0.047, # Taxa de câmbio para Iene Japonês
    "real": 1.0 # Taxa de câmbio para Real Brasileiro (base para conversão)
}

print("Escolha a moeda de origem:") # Exibe as opções de moeda para o usuário
print("1 - Dólar (USD)")
print("2 - Euro (EUR)")
print("3 - Libra Esterlina (GBP)")
print("4 - Iene Japonês (JPY)")
print("5 - Real Brasileiro (BRL)")

while True: # Loop para garantir que o usuário digite uma opção válida
    escolha_moeda = input("Digite o número da moeda de origem: ") # Solicita a escolha da moeda de origem
    if escolha_moeda in ['1', '2', '3', '4', '5']: # Verifica se a escolha é válida
        break # Sai do loop se a escolha for válida
    else:
        print("Escolha inválida. Digite um número de 1 a 5.") # Mensagem de erro para escolha inválida

moedas = { # Dicionário para mapear a escolha do número para o nome da moeda
    '1': 'dolar',
    '2': 'euro',
    '3': 'libra',
    '4': 'iene',
    '5': 'real'
}

moeda_origem = moedas[escolha_moeda] # Obtém o nome da moeda com base na escolha do usuário

while True: # Loop para garantir que o usuário digite um valor numérico
    try:
        valor = float(input(f"Digite o valor em {moeda_origem.capitalize()}: ").replace(",", ".")) # Solicita o valor a ser convertido, tratando vírgulas como pontos decimais
        break # Sai do loop se o valor for válido
    except ValueError: # Captura o erro se o valor não for numérico
        print("Valor inválido. Digite um número.") # Mensagem de erro para valor inválido


if moeda_origem == "real": # Verifica se a moeda de origem é Real
    valor_em_real = valor # Se for Real, o valor em Real é o próprio valor digitado
else: # Se a moeda de origem não for Real
    if moeda_origem in taxas_de_cambio: # Verifica se a moeda está nas taxas de câmbio
        valor_em_real = valor * taxas_de_cambio[moeda_origem] # Calcula o valor em Real multiplicando pelo fator de conversão
    else:
        valor_em_real = 0 # Define o valor em Real como 0 se a moeda não for encontrada nas taxas (pode ser tratado de forma diferente, como uma mensagem de erro)

print(f"O valor de {valor:.2f} {moeda_origem.capitalize()} é equivalente a {valor_em_real:.2f} Reais (BRL).") # Exibe o resultado da conversão formatado com duas casas decimais
```

```python
jogador1 = input("jogador1, jogue pedra, papel ou tesoura: ").lower()
jogador2 = input("jogador2, jogue pedra, papel ou tesoura: ").lower()

if jogador1 == jogador2:
    print("Empate!")
elif jogador1 == "pedra":
    if jogador2 == "papel":
        print("O jogador 2 venceu a partida!")
    elif jogador2 == "tesoura":
        print("O jogador 1 venceu a partida")
elif jogador1 == "papel":
    if jogador2 == "tesoura":
        print("O jogador 2 venceu a partida!")
    elif jogador2 == "pedra":
        print("O jogador 1 venceu a partida!")
elif jogador1 == "tesoura":
    if jogador2 == "pedra":
        print("O jogador 2 venceu a partida!")
    elif jogador2 == "papel":
        print("O jogador 1 venceu a partida!")
else:
  print("jogo inválido!")
```
