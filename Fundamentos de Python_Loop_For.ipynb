# **LAÇO PARA (LOOP FOR)**

Chamado de “Laço Contador”. Este tipo de laço é empregado melhor em situações em que o programador sabe de antemão a quantidade de voltas que serão executadas.

Neste caso é mais importante contar as "voltas" do loop do que verificar as condições.

```python
for <contador> in range(<início>, <fim>, <incremento>):
  <Bloco de repetição>
```

```python
for i in range(100):
    # do_something()
    pass
```

```python
for i in range(10):
   print("O valor de i é atualmente", i)
```

```python
# A invocação da função range() pode ser equipada com dois argumentos, não apenas um:

for i in range(2, 8):
    print("O valor de i é atualmente", i)
```

```python
# A função range() também pode aceitar três argumentos - dê uma olhada no código no editor.

for i in range(2, 8, 3):
  print("O valor de i é atualmente", i)
```

```python
# Nota: se o conjunto gerado pela função range() estiver vazio, o loop não executará seu corpo.

for i in range(1, 1):
    print("O valor de i é atualmente", i)
```

```python
# Range decrescente

for i in range(20,-1,-1):
  print("O valor de i é atualmente", i)
```

**Loop FOR com as  instruções break e continue**

```python
print("The break instrução:")
for i in range(1, 6):
    if i == 3:
        break
    print("Dentro do laço.", i)
print("Fora do loop.")
```

```python
print("The continue instrução:")
for i in range(1, 6):
    if i == 3:
        continue
    print("Dentro do laço.", i)
print("Fora do loop.")
```

```python
# Strings também são sequências

for caracter in 'Python é uma linguagem de programação divertida!':
    print (caracter)
```

**EXERCÍCIO 1:** Mississipi

A palavra Mississipi, além de nomer um rio nos EUA, também é usada para uma finalidade ligeiramente diferente: contar segundos.

A ideia por trás disso é que adicionar a palavra Mississipi a um número ao contar segundos em voz alta faz com que soem mais perto do relógio e, portanto, "um Mississipi, dois Mississipi, três Mississipi" levará aproximadamente três segundos reais! É frequentemente usado por crianças que brincam de esconde-esconde para garantir que o candidato faça uma contagem honesta

Sua tarefa é muito simples aqui: escreva um programa que use um loop FOR para contar (e imprimir) até cinco . Depois de contar até cinco, o programa deve imprimir na tela a mensagem final "Pronto ou não, aqui vou eu!"



```python
import time

# Contagem de 1 a 5 com "Mississipi"
for i in range(1, 6):
    # print(i, "Mississipi")
    print(f"{i} Mississipi")
    time.sleep(1)

# Mensagem final
print("Pronto ou não, aqui vou eu!")
```

**EXERCÍCIO 4:** Faça um programa para percorrer os números ímpares entre 1 e 500 e mostre quantos são os múltiplos de 3 entre eles e some-os.

```python
#somas dos impares multiplos 3 entre 1 e 500
#Mostre também quantos foram somados

soma = 0
qtd = 0

for contador in range(1,500,2):
  if contador % 3 == 0:
    qtd += 1
    soma = soma + contador
print("Soma dos {} numeros impares = {}".format(qtd, soma))
print(f"Soma dos {qtd} numeros impares = {soma}")
print("Soma dos",  qtd, "numeros impares = ", soma)
```
